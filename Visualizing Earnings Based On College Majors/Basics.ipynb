# Analysing Popular Majors Data

This note will go through the numbers in the dataset provided [here](https://github.com/fivethirtyeight/data/tree/master/college-majors), and analyse them as per its columns.  

The data has the below columns:

1. Rank: Rank by median earnings
2. Major_code: Major code, FO1DP in ACS PUMS
3. Major: Major description
4. Major_category: Category of major from Carnevale et al
5. Total: Total number of people with major
6. Sample_size: Sample size (unweighted) of full-time, year-round ONLY (used for earnings)
7. Men: Male graduates
8. Women: Female graduates
9. ShareWomen: Women as share of total
10. Employed: Number employed (ESR == 1 or 2)
11. Full_time: Employed 35 hours or more
12. Part_time: Employed less than 35 hours
13. Full_time_year_round: Employed at least 50 weeks (WKW == 1) and at least 35 hours (WKHP >= 35)
14. Unemployed: Number unemployed (ESR == 3)
15. Unemployment_rate: Unemployed / (Unemployed + Employed)
16. Median: Median earnings of full-time, year-round workers
17. P25th: 25th percentile of earnings
18. P75th: 75th percentile of earnings
19. College_jobs: Number with job requiring a college degree
20. Non_college_jobs: Number with job not requiring a college degree
21. Low_wage_jobs: Number in low-wage service jobs

## Preparation:

```python
#Input variables
fileName = 'recent-grads.csv'
```

```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

#Jupyter magic
%matplotlib inline
```

```python
#Reading dataset
recent_grads = pd.read_csv(fileName)
```

## Selected Columns

1. Rank - Rank by median earnings (the dataset is ordered by this column).
2. Major_code - Major code.
3. Major - Major description.
4. Major_category - Category of major.
5. Total - Total number of people with major.
6. Sample_size - Sample size (unweighted) of full-time.
7. Men - Male graduates.
8. Women - Female graduates.
9. ShareWomen - Women as share of total.
10. Employed - Number employed.
11. Median - Median salary of full-time, year-round workers.
12. Low_wage_jobs - Number in low-wage service jobs.
13. Full_time - Number employed 35 hours or more.
14. Part_time - Number employed less than 35 hours.

## Cleaning

```python
#dropping rows with missing values
raw_data_count = len(recent_grads)
print('Raw data count:', raw_data_count)

recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print('Cleaned data count:', cleaned_data_count)
```
